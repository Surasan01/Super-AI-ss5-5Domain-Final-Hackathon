```python
import pandas as pd
import os
import numpy as np
from datetime import timedelta
```

```python
attendance_df= pd.read_csv('attendance.csv')
train_df = pd.read_csv('train.csv')
submission = pd.read_csv('submission.csv')
```

```python
attendance_df
```

```python
attendance_df['datetime'] = attendance_df['datetime'].sort_values()
```

```python
attendance_df
```

```python
attendance_df = attendance_df.merge(train_df, on='user_id', how='left')
```

```python
attendance_df
```

```python
attendance_df['house'].isnull().sum()
```

```python
attendance_df['house'] = attendance_df['house'].ffill()
```

```python
attendance_df = attendance_df.sort_values('datetime')
attendance_df
```

```python
attendance_df = attendance_df.groupby('user_id')['house'].agg(lambda x: x.mode()[0] if not x.mode().empty else None)
```

```python
sub = pd.DataFrame({
    'user_id' : id,
    'house' : p
})
```
