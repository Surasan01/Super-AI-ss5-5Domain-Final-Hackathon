```python
import opendatasets as od
import pandas as pd
from autogluon.tabular import TabularDataset, TabularPredictor

dataset_url = 'https://www.kaggle.com/competitions/individual-test-temperature-prediction'

od.download(dataset_url)

data_dir = '/content/individual-test-temperature-prediction'

train_data = pd.read_csv(f'{data_dir}/IOT_Train.csv')
test_data = pd.read_csv(f'{data_dir}/IOT_Test.csv')

train_data.head()
```

```python
train_data['pm10'] = train_data['pm10'].fillna(median10)
```

```python
mediant = train_data['temp'].median()
mediant
```

```python
train_data['temp'] = train_data['temp'].fillna(mediant)
```

```python
test_ids = test_data['id']

train_data_ag = TabularDataset(train_data)
test_data_ag = TabularDataset(test_data)
```

```python
train_data_ag['mac'].unique()
```

```python
label = 'temp'
time_limit = 7000

predictor = TabularPredictor(
    label=label,
    eval_metric='root_mean_squared_error',
    path='ag_Temperature_Prediction'
)

predictor.fit(
    train_data_ag,
    time_limit=time_limit,
    presets='best_quality'
)
```

```python
# from google.colab import drive
# drive.mount('/content/drive')
```

```python
predictor.leaderboard(train_data_ag, silent=True)
```

```python

predictions = predictor.predict(test_data_ag)


submission = pd.DataFrame({
    'id': test_ids,
    'temp': predictions
})

submission.head()

# submission.to_csv('submission_Temperature_Prediction_1.csv', index=False)

# submission.to_csv('submission_Temperature_Prediction_3.csv', index=False)

submission.to_csv('/content/drive/MyDrive/submission_Temperature_Prediction_4.csv', index=False)
```
