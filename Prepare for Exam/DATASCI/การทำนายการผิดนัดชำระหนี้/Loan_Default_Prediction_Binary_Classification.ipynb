```python
import opendatasets as od
import pandas as pd
import os

# ระบุ URL ของชุดข้อมูลบน Kaggle
dataset_url = 'https://www.kaggle.com/competitions/home-credit-default-risk'

# ดาวน์โหลดข้อมูล (ระบบจะถามหา Kaggle Username และ Key)
od.download(dataset_url)

# กำหนด Path หลักของข้อมูลที่ดาวน์โหลดมา
data_dir = '/content/home-credit-default-risk'
```

```python
# สร้าง Dictionary เพื่อเก็บ DataFrames ของข้อมูล Training
# เราจะใช้ตารางหลัก และตารางเสริมที่สำคัญ
train_data = {
    'application_train': pd.read_csv(f'{data_dir}/application_train.csv'),
    'bureau': pd.read_csv(f'{data_dir}/bureau.csv'),
    'previous_application': pd.read_csv(f'{data_dir}/previous_application.csv'),
}

# สร้าง Dictionary สำหรับข้อมูล Test ที่มีโครงสร้างเหมือนกัน
test_data = {
    'application_test': pd.read_csv(f'{data_dir}/application_test.csv'),
    'bureau': pd.read_csv(f'{data_dir}/bureau.csv'),
    'previous_application': pd.read_csv(f'{data_dir}/previous_application.csv'),
}

# แสดงขนาดของข้อมูลในตาราง application_train
print("ขนาดของข้อมูล:")
print(train_data['application_train'].shape)

# แสดงตัวอย่างข้อมูล 5 แถวแรก
print("\nตัวอย่างข้อมูล:")
train_data['application_train'].head()
```

```python
from autogluon.tabular import TabularDataset, TabularPredictor

# แปลง Dictionary ของ DataFrame ให้อยู่ในรูปแบบ TabularDataset
# ซึ่งเป็นรูปแบบข้อมูลที่ AutoGluon เข้าใจ
# train_data = TabularDataset(train_data)
# test_data = TabularDataset(test_data)

# กำหนดความสัมพันธ์ระหว่างตาราง
# (ชื่อตารางหลัก, 'Key_หลัก'), (ชื่อตารางรอง, 'Key_รอง')
relationships = [
    ('application_train', 'SK_ID_CURR', 'bureau', 'SK_ID_CURR'),
    ('application_train', 'SK_ID_CURR', 'previous_application', 'SK_ID_CURR'),
]
```

```python
# รวมตารางข้อมูลเข้าด้วยกันสำหรับการฝึก
# เริ่มจากตารางหลัก application_train
train_merged = train_data['application_train'].copy()

# รวมข้อมูล bureau โดยการสร้าง aggregated features
bureau_agg = train_data['bureau'].groupby('SK_ID_CURR').agg({
    'DAYS_CREDIT': ['min', 'max', 'mean'],
    'CREDIT_DAY_OVERDUE': ['max', 'mean'],
    'AMT_CREDIT_MAX_OVERDUE': ['max', 'mean'],
    'AMT_CREDIT_SUM': ['sum', 'mean'],
    'AMT_CREDIT_SUM_DEBT': ['sum', 'mean']
}).reset_index()

# ปรับชื่อคอลัมน์
bureau_agg.columns = ['SK_ID_CURR'] + [f'BUREAU_{col[0]}_{col[1]}' for col in bureau_agg.columns[1:]]

# รวมเข้ากับตารางหลัก
train_merged = train_merged.merge(bureau_agg, on='SK_ID_CURR', how='left')

# รวมข้อมูล previous_application
prev_app_agg = train_data['previous_application'].groupby('SK_ID_CURR').agg({
    'AMT_ANNUITY': ['min', 'max', 'mean'],
    'AMT_APPLICATION': ['min', 'max', 'mean'],
    'AMT_CREDIT': ['min', 'max', 'mean'],
    'AMT_DOWN_PAYMENT': ['min', 'max', 'mean'],
    'AMT_GOODS_PRICE': ['min', 'max', 'mean'],
    'RATE_DOWN_PAYMENT': ['min', 'max', 'mean']
}).reset_index()

# ปรับชื่อคอลัมน์
prev_app_agg.columns = ['SK_ID_CURR'] + [f'PREV_APP_{col[0]}_{col[1]}' for col in prev_app_agg.columns[1:]]

# รวมเข้ากับตารางหลัก
train_merged = train_merged.merge(prev_app_agg, on='SK_ID_CURR', how='left')

# กำหนดคอลัมน์เป้าหมาย
label = 'TARGET'
# กำหนดเวลาในการฝึก (หน่วยเป็นวินาที), 600 วินาที = 10 นาที
# สำหรับการแข่งขันจริง อาจเพิ่มเป็น 3600 (1 ชั่วโมง) หรือมากกว่า
time_limit = 600

# สร้าง Predictor
predictor = TabularPredictor(
    label=label,
    eval_metric='roc_auc',
    path='ag_models_loan_default' # โฟลเดอร์สำหรับเซฟโมเดล
)

# เริ่มฝึกโมเดล!
predictor.fit(
    train_merged,
    time_limit=time_limit,
    presets='best_quality' # ตั้งค่าให้เน้นคุณภาพสูงสุด (อาจใช้เวลามากขึ้น)
)
```

```python
# แสดง Leaderboard ของโมเดลทั้งหมด
# โมเดลที่มี score_val สูงที่สุดคือโมเดลที่ดีที่สุด
predictor.leaderboard()
```

```python
# Combine dataframes for testing
# Start with the main application_test table
test_merged = test_data['application_test'].copy()

# Combine bureau data by creating aggregated features
bureau_agg = test_data['bureau'].groupby('SK_ID_CURR').agg({
    'DAYS_CREDIT': ['min', 'max', 'mean'],
    'CREDIT_DAY_OVERDUE': ['max', 'mean'],
    'AMT_CREDIT_MAX_OVERDUE': ['max', 'mean'],
    'AMT_CREDIT_SUM': ['sum', 'mean'],
    'AMT_CREDIT_SUM_DEBT': ['sum', 'mean']
}).reset_index()

# Adjust column names
bureau_agg.columns = ['SK_ID_CURR'] + [f'BUREAU_{col[0]}_{col[1]}' for col in bureau_agg.columns[1:]]

# Merge with the main table
test_merged = test_merged.merge(bureau_agg, on='SK_ID_CURR', how='left')

# Combine previous_application data
prev_app_agg = test_data['previous_application'].groupby('SK_ID_CURR').agg({
    'AMT_ANNUITY': ['min', 'max', 'mean'],
    'AMT_APPLICATION': ['min', 'max', 'mean'],
    'AMT_CREDIT': ['min', 'max', 'mean'],
    'AMT_DOWN_PAYMENT': ['min', 'max', 'mean'],
    'AMT_GOODS_PRICE': ['min', 'max', 'mean'],
    'RATE_DOWN_PAYMENT': ['min', 'max', 'mean']
}).reset_index()

# Adjust column names
prev_app_agg.columns = ['SK_ID_CURR'] + [f'PREV_APP_{col[0]}_{col[1]}' for col in prev_app_agg.columns[1:]]

# Merge with the main table
test_merged = test_merged.merge(prev_app_agg, on='SK_ID_CURR', how='left')
```

```python
# โหลด ID ของข้อมูล test เพื่อใช้ในไฟล์ submission
test_ids = test_merged['SK_ID_CURR']

# ทำนายความน่าจะเป็นของ Class 1 (ผิดนัดชำระหนี้)
# เราส่งข้อมูล test ทั้งหมดเข้าไปพร้อมกันได้เลย
predictions = predictor.predict_proba(test_merged)

# predictions จะเป็น DataFrame ที่มี 2 คอลัมน์ (ความน่าจะเป็นของ class 0 และ 1)
# เราต้องการแค่คอลัมน์ของ class 1
submission_preds = predictions[1]

# สร้าง DataFrame สำหรับไฟล์ submission
submission = pd.DataFrame({'SK_ID_CURR': test_ids, 'TARGET': submission_preds})

# แสดงตัวอย่าง 5 แถวแรกของไฟล์ submission
print("ตัวอย่างไฟล์ Submission:")
submission.head()

# บันทึกเป็นไฟล์ CSV
submission.to_csv('submission.csv', index=False)

print("\nสร้างไฟล์ submission.csv เรียบร้อยแล้ว!")
```
